```python
def convertRange(x):

    temp = x.split('-')

    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2

    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers_sqft(df):
  if df.empty:
    print("DataFrame is empty, cannot perform outlier removal.")
    return pd.DataFrame()

  df_output = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)

    gen_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
    df_output = pd.concat([df_output, gen_df], ignore_index = True)
  return df_output
df = remove_outliers_sqft(df)
```

```python
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean' : np.mean(bhk_df.price_per),
          'std' : np.std(bhk_df.price_per_sqft),
          'count' : bhk_df.shape[0]
      }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk - 1)
      if stats and stats['count'] > 5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
  return df.drop(exclude_indices, axis = 'index')
```

```python
X = df.drop(columns = ['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Appling Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False), ['location']), remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipelr = make_pipeline(column_trans, scaler, lr)
```

```python
pipelr.fit(X_train, y_train)
```

```python
y_pred_lr = pipelr.predict(X_test)
```

```python
r2_score = r2_score(y_test, y_pred_lr)
```

```python
print(r2_score)
```

## Applying Lasso

```python
lso = Lasso()
```

```python
pipel = make_pipeline(column_trans, scaler, lso)
```

```python
pipel.fit(X_train, y_train)
```

```python
y_pred_l = pipel.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred_l)
```

```python
print(r2_score)
```

## Applying Ridge

```python
rge = Ridge()
```

```python
piper = make_pipeline(column_trans, scaler, rge)
```

```python
piper.fit(X_train, y_train)
```

```python
y_pred_r = piper.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred_r)
```

```python
print(r2_score)
```
